### Observation : Looks like few of the numerical variable have some outliers. Lets cap them to  Q3+1.5*IQR and Q1-1.5*IQR

```python
# for col in df[num_cols].columns:
#     df[num_cols][col][df[num_cols][col] <= df[num_cols][col].quantile(0.05)] = df[num_cols][col].quantile(0.05)
#     df[num_cols][col][df[num_cols][col] >= df[num_cols][col].quantile(0.95)] = df[num_cols][col].quantile(0.95)
```

```python
for col in num_cols:
    # First quartile (Q1)
    Q1 = np.percentile(df[col], 25, interpolation = 'midpoint')
    # Third quartile (Q3)
    median = np.percentile(df[col], 50, interpolation = 'midpoint')
    # Third quartile (Q3)
    Q3 = np.percentile(df[col], 75, interpolation = 'midpoint')
    # Interquaritle range (IQR)
    IQR = Q3 - Q1
    upper_limit = Q3 + 1.5 * IQR
    lower_limit = Q1 - 1.5 * IQR
    
    df[col][df[col] <= lower_limit] = lower_limit
    df[col][df[col] >= upper_limit] = upper_limit
    print("{} : Q1:{}, median:{}, Q3:{}, upper_limit : {}, lower_limit: {}".format(col,Q1,median,Q3,upper_limit,lower_limit))
    #print("upper limit : {} and lower limit : {}".format(upper_limit,lower_limit))
```

### ANOVA testing on categorical variables to select only significant categorical variable

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols                      
from statsmodels.stats.anova import anova_lm          # For n-way ANOVA
```

```python
cat_col_todrop = []                                  ## iterate to all categorical variables  
for col in cat_cols:
    formula = 'SalePrice ~ {}'.format(col)
    model = ols(formula,data=df).fit()
    aov_table = anova_lm(model,typ=2)
    print(aov_table, '\n')
    if aov_table['PR(>F)'][0] > .05:                  ## if p-value is greater than alpha(.05) then drop it from further processing.  
        cat_col_todrop.append(col) 
print(cat_col_todrop)
df.drop(columns=cat_col_todrop,axis=1,inplace=True)   ## drop the columns from dataframe   
```

### 4. Building the statsmodel with all variable first

```python
import statsmodels
import statsmodels.api as sm
```

```python
### Add a constant variable as OLS model consider y intercept 0 by default
X_train_sm = sm.add_constant(X_train)
```

```python
X_train_sm.head(2)
```

```python
def build_ols_model(y,X):
    
    X = sm.add_constant(X)    ## Add a constant value before training

    ### build the instance of ols model
    model = sm.OLS(y,X)
    
    ## fit the model
    lr_model = model.fit()
    
    ## model summary
    return lr_model, X    
```

```python
# ## training the model with all features initially
lr_model, X_train_sm = build_ols_model(y_train,X_train_sm)
lr_model.summary()
```

### Observation: R2 value (with all features) is 95.9

```python
### Print the Pvalue and Coefficients from the summary for feature selection
from statsmodels.stats.outliers_influence import variance_inflation_factor
def report_summary(model):
    model_summary =  model.summary2().tables[1][['Coef.','P>|t|']].reset_index().rename(columns={'index':'feature'})
    model_summary['vif'] = [variance_inflation_factor(X_train_sm.values,i).round(2) for i in range(len(X_train_sm.columns))]
    return model_summary
```

```python
## get the pvalue and vif for all features
lr_model_summary = report_summary(lr_model)

## get the list of insignificant features having pvalue > .05(significant level) and vif > 5
insignificant_feature_1 = list(lr_model_summary[(lr_model_summary['P>|t|'] > .05) & (lr_model_summary['vif'] > 5)]['feature'])
len(insignificant_feature_1)
```

```python
lr_model_summary.head()
```

```python
### Drop insignificant variable and retrain the model
X_train_sm.drop(columns=insignificant_feature_1,axis=1,inplace=True)
print('shape of X_train_sm:', X_train_sm.shape)

## training the model with new features 
lr_model, X_train_sm = build_ols_model(y_train,X_train_sm)
lr_model.summary()

## get the pvalue and vif for all features
lr_model_summary = report_summary(lr_model)

## get the list of insignificant features having pvalue > .05(significant level) 
insignificant_feature_2 = list(lr_model_summary[(lr_model_summary['P>|t|'] > .05)]['feature'])
len(insignificant_feature_2)
```

## Ridge Regularization
### Regularization penalizes the low significent coefficinets to prevent overfitting. 

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lm_ridge = Ridge()
lambdas = [0.001, 0.01, 0.1, 1, 4, 5, 6, 7, 10,15, 20, 50, 100, 1000]
parameters= {'alpha': lambdas}

lm_ridge_grd = GridSearchCV(estimator = lm_ridge, param_grid=parameters, cv=10)

lm_ridge_grd.fit(X_train, y_train)
```

```python
print(' Best model parameter : {}  \n Best Avg_cv score : {}'.format(lm_ridge_grd.best_params_, lm_ridge_grd.best_score_))
```

```python
### Retrain the Ridge with best estimator value ( i.e. with alpha = 10)
lm_ridge_best = lm_ridge_grd.best_estimator_
lm_ridge_best.fit(X_train,y_train)

### predict on train dataset
X_train_pred_r = lm_ridge_best.predict(X_train)

### R2 value on train dataset
print('R2 value on train dataset using Ridge regularization:', r2_score(X_train_pred_r,y_train).round(2))
print('RMSE value on train dataset using Ridge regularization:', np.sqrt(mean_squared_error(X_train_pred_r,y_train)).round(2))
### predict on test dataset
X_test_pred_r = lm_ridge_best.predict(X_test)

### R2 value on test dataset
print('R2 value on train dataset using Ridge regularization:', r2_score(X_test_pred_r,y_test).round(2))
print('RMSE value on test dataset using Ridge regularization:', np.sqrt(mean_squared_error(X_test_pred_r,y_test)).round(2)   )
```
